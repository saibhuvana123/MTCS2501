```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM



model_name = 'Helsinki-NLP/opus-mt-mul-en'

tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

print(f"Model '{model_name}' loaded successfully!")
```

```python
def translate_sanskrit_to_english(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True, max_length=512)
    outputs = model.generate(**inputs, max_new_tokens=512)
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text

sanskrit_sloka = "यदा यदा हि धर्मस्य ग्लानिर्भवति भारत। अभ्युत्थानमधर्मस्य तदात्मानं सृजाम्यहम्॥"

english_translation = translate_sanskrit_to_english(sanskrit_sloka)

print("Sanskrit Sloka:", sanskrit_sloka)
print("English Translation:", english_translation)
```
